```python
import re

with open ("/Users/Nils/Projekte/Satz-Reduktion/GIT/Datensatz/10000 Sätze aus der Tm2018 Datenbank", "r") as file:
    text = file.read().replace('\n',' ')

    # Temporale Präpositionen

    # ersetzt Zeitpunkte mit Am/An Tage, Datum, Tageszeiten und Feiertage
    text = re.sub(r"((Am|An|am|an)\s(\b\w*\b\s)*((\d{1,2}\.(\d{1,2}|\s(Januar|Februar|März|April|Mai|Juni|Juli|August|September|Oktober|November|Dezember))((\.)?\d{0,4}))|(Montag|Dienstag|Mittwoch|Donnerstag|Freitag|Samstag|Sonntag)|(Morgen|Mittag|Abend|Nacht|Vormittag|Nachmittag)|(Weihachten|Ostern|Muttertag|Pfingsten|Buß- und Bettag|Neujahr|Tag der deutschen Einheit|Tag der Arbeit|Christi Himmelfahrt)))", "IRGENDWANN", text)

    # ersetzt Zeitpunkte mit Im/In
    text = re.sub(r"(Im|In|im|in)\s(\b\w*\b\s)*((\d{2}er)|(Wochen)|(Zeit)|(Januar|Februar|März|April|Mai|Juni|Juli|August|September|Oktober|November|Dezember)|(Frühling|Sommer|Herbst|Winter))", "IRGENDWANN", text)

    # ersetzt Zeitpunkt Uhrzeit oder Jahreszahl mit Um
    text = re.sub(r"(Um|um|Punkt|punkt)\s(\d{1,4}|\d{1,2}(:|\.)(\s)?(\d{1,2})?)+\s(Uhr)?", "IRGENDWANN", text)

    # ersetzt Zeitpunkt mit Vor + Dativ
    text = re.sub(r"(Vor|vor)\s(dem|den)\s(\w+\b)*", "IRGENDWANN", text)
    
    # ersetzt Zeitpunkt mit ab
    text = re.sub(r"(Ab|ab)\s((\d{1,2}(:|\.)?(\s)?(\d{1,2})?)+\s(Uhr\s)?)|(Januar|Februar|März|April|Mai|Juni|Juli|August|September|Oktober|November|Dezember)\s(\s\d{1,4}\s)?|(\b\w*\b\s)*(Frühling|Sommer|Herbst|Winter)|(\d{4}\/\d{4}\s)|(Jahre\s)","Irgendwann", text)

    # ersetzt Jahreszahlen
    text = re.sub(r"(\d{4})", "IRGENDWANN", text)

    # Temporale Adverbien
    text = re.sub(r"(montags|dienstags|mittwochs|donnerstags|freitags|samstags|sonntags)","IRGENDWANN", text)

    print(text)
```
